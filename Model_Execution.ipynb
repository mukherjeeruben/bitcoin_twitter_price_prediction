```python
def average(lst):
    return sum(lst) / len(lst)
```

```python
df = pd.DataFrame(columns=["date","close_price"])
```

```python
import math
```

```python
previous_price = 4.39
size = len(price_df)
bitcoin_df = {}
i = 0
while i < size and i+60 < size:
    price = []
    for j in range(i, i+60):
        if not math.isnan(price_df["Close"][j]):
            price.append(price_df["Close"][j])
    if len(price) != 0:
        close_price = round(average(price), 2)
    else:
        close_price = previous_price
    data = {"date": price_df["date"][i], "close_price": previous_price}
    previous_price = close_price
    df = df.append(data, ignore_index=True)
    i = i + 60
```
